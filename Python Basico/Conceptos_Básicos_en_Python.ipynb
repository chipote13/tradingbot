## Funciones en Python

```python
def suma(a,b):
    return a+b
a = 40
b = 10
suma(a,b)
suma(3,4)
```

```python
def resto(a,b):
    return a%b
resto(5,5)
```

```python
def cuenta(c):
    return c.count("Hola")
cuenta("Hola don pepito, Hola don jose,Hola a todos")
```

```python
# 4! =4*3*2
def recursion(x):
    if x == 1:
        return 1
    else:
        return (x * recursion(x-1))

recursion(4)
```

# Clases en Python

```python
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    def esmayoredad(self):
        if self.edad>= 18:
            return True
        else:
            return False
    def setedad(self,edad):
        self.edad = edad
    def setnombre(self,nombre):
        self.nombre = nombre
    def getedad(self):
            return self.edad
    
p1 = Persona("Kevin", 26)
print(p1.nombre)
print(p1.edad)
print(p1.esmayoredad())
p1.setedad(13)
print(p1.getedad())
p1.esmayoredad()
```
