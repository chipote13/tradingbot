# Uso de Loc e iloc Para filtrar en DataFrames


```
from seaborn import load_dataset
df = load_dataset("tips")
df.head()
```

# El método iloc se utiliza en los DataFrames para seleccionar los elementos en base a su ubicación. Su sintaxis es data.iloc[filas,columnas]

```
df.iloc[0]
```

```
df.iloc[-1]
```

```
df.iloc[:, 0]
```

```
df.iloc[:, -1]
```

**Selección de Multiples Columnas**


```
df.iloc[0:10] #filas
```

```
df.iloc[:, 0:4] #columnas
```

```
df.iloc[[0,1,-1]] #primera , segunda y cuarta fila
```

```
df.iloc[:,[0,3,1]] #primera , segunda y cuarta columna
```

```
df.loc[:, 'smoker'] # Columna tip
```

```
df.loc[:, ['total_bill', 'sex', 'day']] # Columnas tip, sex y day
```

# Filtros en DataFrames

```
is_male = df.loc[:, 'sex'] == 'Male'
df_male = df.loc[is_male]
df_male.head(10)
```

```
total_up_10 =df.loc[:, 'total_bill'] >9
df_male = df.loc[total_up_10]
df_male.head(20)
```

```
fumador = df.loc[:, 'smoker'] =='Yes'
df_fumador = df.loc[fumador]
df_fumador.head(20)
```
